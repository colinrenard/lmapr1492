```python
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("rtIvjK8boEIrBtPMF") as m:

    # Crée la structure pour le matériau considéré
    # structure = m.get_structure_by_material_id("mp-1011695", conventional_unit_cell=True)
    structure = m.get_structure_by_material_id("mp-1011695")
```

La maille considérée est la cellule primitive.

```python
xrd = XRDCalculator(wavelength=1.5406)
pattern = xrd.get_pattern(structure)

n_peaks = 3 # nombre de pics que l'on souhaite analyser

xrd.show_plot(structure)
# xrd.show_plot(structure, annotate_peaks=False)
```

# Analyse des trois premiers pics de diffraction

```python
diff_peak = []

for i in range(n_peaks):
    theta = pattern.x[i]
    intensity = pattern.y[i]
    indexes = pattern.hkls[i]
    peak_data = [theta, intensity, indexes]
    diff_peak.append(peak_data)
```

```python
print("Premier pic de diffraction : ")
print("Angle : "+str(diff_peak[0][0]))
print("Intensité : "+str(diff_peak[0][1]))
print("Indices hkl : "+str(diff_peak[0][2][0]['hkl']))
print("Multiplicité : "+str(diff_peak[0][2][0]['multiplicity']))
```

```python
print("Deuxième pic de diffraction : ")
print("Angle : "+str(diff_peak[1][0]))
print("Intensité : "+str(diff_peak[1][1]))
print("Indices hkl : "+str(diff_peak[1][2][0]['hkl']))
print("Multiplicité : "+str(diff_peak[1][2][0]['multiplicity']))
```

```python
print("Troisième pic de diffraction : ")
print("Angle : "+str(diff_peak[2][0]))
print("Intensité : "+str(diff_peak[2][1]))
print("Indices hkl : "+str(diff_peak[2][2][0]['hkl']))
print("Multiplicité : "+str(diff_peak[2][2][0]['multiplicity']))
```

On observe bien les valeurs déterminées sur le graphe ci-dessus. On note que l'angle retourné est bien $2\theta$.
